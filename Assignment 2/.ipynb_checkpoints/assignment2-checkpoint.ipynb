````python
# YOUR CODE HERE

def net_input(X, W):    
    ''' Computes the matrix product between X and w. Note that
    X is assumed not to contain a bias/intercept column.'''
    return np.dot(X, W[1:]) + W[0]   
                                        

def compute_error(y, X, W):
    return y - net_input(X, W)



def update_weight(X, y, W, eta):
    '''Updates the weight vector W using gradient descent.'''
    error = compute_error(y, X, W)
    
    # Compute the gradient
    gradient = np.dot(X.T, error)

    # Update the weights
    W[1:] += eta * gradient  # Adjusted to add the gradient (descent step)
    W[0] += eta * np.sum(error)  # Update the bias term with sum of errors
    
    return w


````

> **Ex. 12.2.3:**
Run a Lasso regression using the Pipeline from `Ex 12.2.2` which is given in the code cell below. In the outer loop searching through the 12 lambda values as specified below. 
In the inner loop make *5 fold cross validation* on the selected model and store the average MSE for each fold. Which lambda, from the selection below, gives the lowest test MSE?
 ```python 
lambdas =  np.logspace(-4, 4, 12)
```
 *Hint:* `KFold` in `sklearn.model_selection` may be useful.

````python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import Lasso

# Pipeline from Ex. 12.2.2
pipe_lasso = make_pipeline(
    PolynomialFeatures(degree=3, include_bias=False),
    StandardScaler(),
    Lasso(random_state=1),
)

# The feature matrix X and target vector y are defined
# in the code cell "Second part of Exercise Set 11: Modelling houseprices".

# Split X, y into development (2/3) and test data (1/3).
X_dev, X_test, y_dev, y_test = train_test_split(X, y, test_size=(1 / 3), random_state=1)

lambdas = np.logspace(-4, 4, 12)

# YOUR CODE HERE
from sklearn.model_selection import KFold
kfolds = KFold(n_splits=5) #5 fold cross-validation.
folds = list(kfolds.split(X_dev, y_dev))

# Outer loop: lambdas
mseCV = []
for lambda_ in lambdas:    
    pipe_lasso.set_params(lasso__alpha=lambda_)
    
    # Inner loop: folds
    mseCV_ = []    
    for train_idx, val_idx in folds:        
        # Train model and compute MSE on test fold           
        X_train, y_train = X_dev.iloc[train_idx], y_dev[train_idx]
        X_val, y_val = X_dev.iloc[val_idx], y_dev[val_idx] 
        pipe_lasso.fit(X_train, y_train)        
        mseCV_.append(mean_squared_error(pipe_lasso.predict(X_val), y_val))    
        
    # Store result    
    mseCV.append(mseCV_) 
    
# Convert to DataFrame
lambdaCV = pd.DataFrame(mseCV, index=lambdas)

#Lowest MSE?
optimal_lambda = lambdaCV.mean(axis=1).nsmallest(1)
print("Optimal lambda: ", optimal_lambda.index[0], "Minimum MSE:", optimal_lambda.iloc[0])
print(lambdaCV)
````

````python
# Import the re module used to construct and use regexes in python
import re
````

````python
TEST_STRINGS = [
    ", Socialdemokratiet.Kl. 09:03(Ordfrer)Christian Rabjerg Madsen (S):Tak for det. For to ", 
    "det og gldelig jul.Kl. 09:12Formanden (Henrik Dam Kristensen):Ja, det varer lidt, ", 
    "nd Poulsen, Venstre.Kl. 09:12Troels Lund Poulsen (V):Tak for det. Tak for", 
    "r Socialdemokratiet?Kl. 09:13Formanden (Henrik Dam Kristensen):Ordfreren, vrsgo.K", 
]

re_find_speakers = re.compile(r"""
    (Kl\.\s\d{2}:\d{2})       # Group 1: Matches the time (e.g., "Kl. 09:13")
    \s*                       # Optional spaces
    (Ordfrer|Formanden)?     # Group 2: Matches the status (Ordfrer or Formanden), optional
    \s*                       # Optional spaces
    ([\(\w+\s\w+\s|\-\b\w*sen\b\)]*) # Group 3: Matches a name ending with "sen", in parentheses or not. Allows spaces, hyphens, etc.
    \s*                       # Optional spaces
    (\(\w\))?                 # Group 4: Matches the party letter in parentheses (e.g., "(S)"), optional
    (:)                       # Group 5: Matches the colon
""", re.VERBOSE)

# Write your regex inside the string

# YOUR CODE HERE
# Test the regex with the provided strings
for string in TEST_STRINGS:
    match = re_find_speakers.search(string)
    print(match)

````

````python
# YOUR CODE HERE

speeches = []
speakers = []

for i in range(len(matches)):
    start_idx, end_idx = matches[i].span()
    speaker = parliament_meeting[start_idx:end_idx]

    # Find the end of the current speaker's speech
    if i < len(matches) - 1:
        next_start_idx = matches[i + 1].span()[0]
    else:
        next_start_idx = len(parliament_meeting)

    # Extract the speech
    speech = parliament_meeting[end_idx:next_start_idx].strip()
    speeches.append(speech)
    speakers.append(speaker)

speech_speaker_dict = dict(zip(speakers, speeches))

for speaker, speech in speech_speaker_dict.items():
    print(f"Speaker: {speaker}\n")
    print(speech)
    print("\n")
    print("-"*50)

    
    
````
