```python
#DSB STATIONER
# KØR IKKE CELLEN FØR #POSTNUMRE FRA TABEL POSTNR1
url_DSB = "https://www.dsb.dk/kundeservice/stationer/" 
response = requests.get(url_DSB, headers=headers)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')

# Find the <div> element that contains the stations
stations_list = soup.find('div', class_='stations-list')

# Extract all <li> elements within that <div>
stations = stations_list.find_all('li')

# Use a list comprehension to create a list of href links
station_links = [station.find('a')['href'] for station in stations if station.find('a')]

# Display the list of href links
#station_links

#full links
full_links = []
for link_end in station_links:
    link_start = "https://www.dsb.dk"
    full_link = link_start+link_end
    full_links.append(full_link)

full_links = full_links[25:]


#HENTER POSTNUMRE FRA DSB-STATIONER

# Function to extract the first 4-digit postal code from a given URL
def get_zipcode(url):
    response = requests.get(url, headers=headers)
    html_content = response.content

    # Parse the HTML content using lxml
    tree = html.fromstring(html_content)

    # Find all <p> elements
    p_elements = tree.xpath('//p')

    # Loop through each <p> element and search for four-digit numbers
    for element in p_elements:
        text = element.text_content().strip()
        # Use regex to find four-digit numbers
        match = re.search(r'\b\d{4}\b', text)
        if match:
            return match.group(0)
    return None


# List to store all found postal codes
postal_codes = []

# Loop through all URLs and get the postal code for each
for link in full_links:
    zipcode = get_zipcode(link)
    if zipcode:
        postal_codes.append(zipcode)

postal_code_counts = Counter(postal_codes)
postal_code_df = pd.DataFrame(postal_code_counts.items(), columns=['Postnummer', 'Stationer'])
postal_code_stations = postal_code_df.groupby('Postnummer')['Stationer'].sum().reset_index()
print(postal_code_stations)
```

```python
#LÆGER
laeger = pd.read_csv('CSV_input/LaegepraksisDK.csv', encoding='ISO-8859-1', delimiter=';') #kilde: https://medcom.dk/standarder/ydere-lokationsnumre/laegepraksis-i-danmark/
alm_laeger = laeger[laeger['Ydertype'].isin([1, 2, 3, 4, 5])] # Dokumentation for at ydertype 1-5 er alm. læger: https://www.dst.dk/da/Statistik/dokumentation/Times/forebyggelsesregistret/ydtyp
alm_laeger = alm_laeger.groupby('Postnr').size().reset_index(name='Antal_laeger')
alm_laeger
```

```python
#Kalder kolonnen med postnr for postal code
alm_laeger.rename(columns={'Postnr': 'Postal code'}, inplace=True)
postal_code_stations.rename(columns={'Postnummer': 'Postal code'}, inplace=True)
```

```python
alm_laeger['Postal code'] = alm_laeger['Postal code'].astype('int32')
postal_code_stations['Postal code'] = postal_code_stations['Postal code'].astype('int32')
```

```python
#Merger læger og statoiner på Postal code
laeger_stationer_postnr_df = pd.merge(alm_laeger, postal_code_stations, on = 'Postal code', how = "outer")
```

```python
laeger_stationer_postnr_df = laeger_stationer_postnr_df.fillna(0)
laeger_stationer_postnr_df['Antal_laeger'] = laeger_stationer_postnr_df['Antal_laeger'].astype('int32')
laeger_stationer_postnr_df['Stationer'] = laeger_stationer_postnr_df['Stationer'].astype('int32')
laeger_stationer_postnr_df = laeger_stationer_postnr_df[laeger_stationer_postnr_df['Postal code'] != 0]
```
